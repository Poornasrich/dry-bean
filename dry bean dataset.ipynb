```python
#dependent and independent columns
x=df.drop('Class',axis=1)
```

```python
y=df['Class']
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder  
la_en=LabelEncoder()
```

```python
y= la_en.fit_transform(df['Class'])
df
```

```python
y
```

```python
#skewness of the data
df.skew()
```

```python
#applying squareroot function to reduce skewness 
p=[]
for i in x.columns:
    b=np.sqrt(x[i])
    X1=pd.DataFrame(b)
    X2=X1.transpose()
    p.append(np.array([b]))
X=pd.DataFrame(np.concatenate(p))

X=X.transpose()
X.columns=x.columns
```

```python
#skewness after applying the sqrt function
X.skew()
```

```python
#handling imbalance data
from imblearn.combine import SMOTEENN
sm=SMOTEENN()
x_sm,y_sm=sm.fit_resample(X,y)
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sm,y_sm,test_size=0.3)
```

```python
#shape of the train data
x_train.shape
```

```python
#standardscaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
#fitting and transform the train data
x_train=scaler.fit_transform(x_train)
```

```python
#transform the test data
x_test=scaler.transform(x_test)
```

```python
#importing confusion matrix,accuracy score
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
#plotting the box plot
j=1
plt.figure(figsize=(20,40))
col = ['Area','Perimeter','MajorAxisLength','MinorAxisLength','AspectRation','Eccentricity','ConvexArea','EquivDiameter','Extent','Solidity','roundness','Compactness','ShapeFactor1','ShapeFactor2','ShapeFactor3','ShapeFactor4']
for i in col:
    plt.subplot(len(col)+1,2,j)
    sns.boxplot(df[i])
    plt.title(i,color='white')
    j=j+1
```
